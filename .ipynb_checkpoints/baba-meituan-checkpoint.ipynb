```python
import requests
import json
import re
cs_url = 'https://club.jd.com/comment/productPageComments.action'
param  = {'callback':'fetchJSON_comment98vv626',
'productId':'4277184',
'score':'0',
'sortType':'5',
'page':0,
'pageSize':'10',
'isShadowSku':'0'}
while(param['page']<10):
    param['page']+=1
    r = requests.get (cs_url, params = param).text
    t = re.search(r'(?<=fetchJSON_comment98vv626\().*(?=\);)',r).group(0)
    x = json.loads(t)
    for i in x['comments']:
        print(i['content'])
```
